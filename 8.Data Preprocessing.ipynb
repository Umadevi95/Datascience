```python
def quanqual(dataset):   #function
        quan=[]
        qual=[]
        for columnName in dataset.columns:
               # print(columnName)
                if(dataset[columnName].dtype=='O'):
                    #print("qual")
                    qual.append(columnName)
                else:
                   # print("quan")
                    quan.append(columnName)
        return quan,qual
```

```python
quan,qual=quanqual(dataset)
```

```python
dataset[quan]
```

```python
#using this to handle missing values 
import numpy as np
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
imp.fit(dataset[quan])
imputed_data=imp.transform(dataset[quan])

#can handle categorical data using mode
```

```python
imputed_data
```

```python
imputed_df=pd.DataFrame(imputed_data,columns=quan)
```
