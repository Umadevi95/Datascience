```python
import pandas as pd
```

```python
dataset=pd.read_csv("Placement.csv")
```

```python
dataset
```

```python
dataset.isnull().sum()    
#no missing values/NaN/Null & outliers before ML&DL
#salary -67 -->Nan values
#relationship logic-->who not placed got no salary
```

```python
#check the below types which suitable for this
#1.replace the missing values with central tendency-mean,median,mode
#2.if many ,entire row delete
#3.respective to the problem statement ,have to replace the missing values
#4.consider this as a semi supervised and create a model,so it can predict the missing values
```

```python
dataset.isna().sum()
```

```python
dataset["salary"].fillna(0,inplace=True)
#inplace=true-->no variable assigned so save it as a final variable
```

```python
dataset.isnull().sum()
```

```python
#delete null values ,entirerow
dataset.dropna(inplace=True)
```

```python
dataset
```

```python
dataset["salary"].fillna(dataset["salary"].mean(),inplace=True)
```

```python
dataset
```

```python
def quanqual(dataset):   #function
        quan=[]
        qual=[]
        for columnName in dataset.columns:
               # print(columnName)
                if(dataset[columnName].dtype=='O'):
                    #print("qual")
                    qual.append(columnName)
                else:
                   # print("quan")
                    quan.append(columnName)
        return quan,qual
```

```python
quan,qual=quanqual(dataset)
```

```python
dataset[quan]
```

```python
#using this to handle many columns missing values 
import numpy as np
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
imp.fit(dataset[quan])#we take quan becoz the dataset having character also so it shows error,
#considering that we take quan for values only
imputed_data_num=imp.transform(dataset[quan])

#can handle categorical data using mode
```

```python
imputed_data_num
```

```python
imputed_df_num=pd.DataFrame(imputed_data_num,columns=quan)
```

```python
imputed_df_num
```

```python
#categorical
import numpy as np
from sklearn.impute import SimpleImputer
imp= SimpleImputer(strategy='most_frequent')
imp.fit(dataset[qual])#we take quan becoz the dataset having character also so it shows error,
#considering that we take quan for values only
imputed_data_cat=imp.transform(dataset[qual])
```

```python
imputed_data_cat
```

```python
imputed_df_cat=pd.DataFrame(imputed_data_cat,columns=qual)
```

```python
preprocessed_data_combined = pd.concat([imputed_df_num, imputed_df_cat], axis=1)
```

```python
preprocessed_data_combined
```

```python
file_name = "preplacement_data.csv"

# Save the preprocessed data to a CSV file
preprocessed_data_combined.to_csv(file_name, index=False)
```
